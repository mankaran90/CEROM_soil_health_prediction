## **Identifying Treatments with More Than Four Cuts per Year in the CEROM Agricultural Experiment**

```python
# Import the necessary libraries
import numpy as np
import pandas as pd
```

```python
# Import the yield dataset
data = pd.read_excel('../data/interim/05_CEROM_Yield_With_Flooding_Related_Columns_2008-24.xlsx', sheet_name='Sheet1')
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# Get the list of unique treatments in the dataset
treatments_list = data['Treatment'].value_counts().reset_index()['Treatment'].to_list()
```

```python
# Display the list of unique treatments
treatments_list
```

```python
# Loop through each treatment and count the number of cuts for each year from 2008 to 2024
for i in treatments_list:
  print('Treatment:',i)
  for j in range(2008,2025):
    print(data[data['Treatment'] == i][data[data['Treatment'] == i]['Year'] == j].shape[0],'number of cuts','for the year',j)
```

```python
# Display the rows for a X.R4.M.E treatment
data_XR4ME = data[data['Treatment'] == 'X.R4.M.E']
```

```python
# Display the rows for a X.R4.M.E treatment
data_XR4ME
```

```python
# Display the rows for X.R4.M.E treatment for the years 2021 to 2023 which have more than 4 cuts in a year
data_XR4ME_2021_to_2023 = data_XR4ME[(data_XR4ME['Year'] >= 2021) & (data_XR4ME['Year'] <= 2023)]
```

```python
# Display the rows for X.R4.M.E treatment for the years 2021 to 2023
data_XR4ME_2021_to_2023
```

## **Conclusion** - Treatment X.R4.M.E has more than 4 cuts per year (4 for four replications) for the years 2021 to 2023 meaning that the crop was harvested not in a single go but many times in that year
